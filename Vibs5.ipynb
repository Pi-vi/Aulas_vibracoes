## Vibração de Modo Normal

$\to$ Um sistema com dois graus de liberdade terá também duas frequências naturais. 

$\to$ Quando ocorre vibração livre com uma destas frequências naturais, existe uma relação definida entre as amplitudes das duas coordenadas e a configuração é referida como modo normal. 

```python
Image('figuras/fig5-3.png',width=600)
```

$m\ddot{x}_1 = -k(x_1 - x_2) -kx_1$

$m \ddot{x}_2 = k(x_1 - x_2) - k x_2$ 

Em equação matricial:

$\left[ \begin{array}{} m && 0 \\ 0 && 2m \end{array}\right] \left[ \begin{array}{} \ddot{x}_1 \\ \ddot{x}_2 \end{array} \right] + \left[ \begin{array}{} 2k && -k \\ -k && 2k \end{array}\right] \left[ \begin{array}{} {x}_1 \\ {x}_2 \end{array} \right] = 0  $

Definimos uma oscilação de modo normal como aquela na qual cada massa é submetida ao movimento harmônico da mesma frequência, passando simultaneamente pela posição de equilíbrio.

$x_1 = A_1 e^{i \omega t} \qquad x_2 = A_2 e^{i \omega t}$

Substituindo nas equações diferenciais: 

$(2k - \omega^2 m) A_1 - k A_2 = 0$

$-k_1 A_1 + (2k - 2 \omega^2 m)A_2 = 0$

$\left[ \begin{array}{} 2k - \omega^2m && -k \\ -k && 2k - 2\omega^2 m \end{array}\right] \left[ \begin{array}{} A_1 \\ A_2 \end{array} \right]$

Fazendo $ \omega^2 = \lambda$, btemos a equação característica:

$\lambda^2 - \left( 3 \dfrac{k}{m} \right) \lambda + \dfrac{3}{2} \left( \dfrac{k}{m} \right)^2 = 0$

As duas raízes desta equação são: 

$\lambda_1 = \left(\dfrac{3}{2} - \dfrac{1}{2} \sqrt{3} \right) \dfrac{k}{m} = 0,634 \dfrac{k}{m}$

$\lambda_2 = \left(\dfrac{3}{2} + \dfrac{1}{2} \sqrt{3} \right) \dfrac{k}{m} = 2,366 \dfrac{k}{m}$

$\omega_1 = \sqrt{\lambda_1} = \sqrt{0,634 \dfrac{k}{m} }$

$\omega_2 = \sqrt{\lambda_2} = \sqrt{2,366 \dfrac{k}{m} }$

Para $\omega_1^2 = 0,634 \dfrac{k}{m}$

$\left( \dfrac{A_1}{A_2} \right)_1 = \dfrac{k}{2k - \omega_1^2 m} = \dfrac{1}{2 - 0,634} = 0,731$

Para $\omega_2^2 = 2,366 \dfrac{k}{m}$

$\left(\dfrac{A_1}{A_2} \right)_2 = \dfrac{k}{2k \omega_2^2 m} = \dfrac{1}{2 - 2,366}^= -2,73$

No primeiro modo de vibração, as duas massas movem-se em fase. 

No segundo, as massas movem-se em oposição de fase, uma em relação à outra.

## Fenômeno de Batimento

Um fenômeno interessante que ocorre quando as frequências naturais de um sistema de dois graus de liberdade são muito próximas

$mL^2 \ddot{\theta}_1 + mgL \theta_1  + k a^2(\theta_1 - \theta_2) = 0$

$mL^2 \ddot{\theta}_2 + mgL \theta_2 - k a^2 (\theta_1 - \theta_2) = 0$



```python
Image('figuras/fig5-5.png',width=600)
```

```python
Image('figuras/fig5-6.png',width=600)
```

$mL^2 \ddot{\theta}_1 + mgL \theta_1 + ka^2(\theta_1 - \theta_2) = 0$

$mL^2 \ddot{\theta}_2 + mgL \theta_2 - k a ^2 ( \theta_1 - \theta_2 ) = 0 $ 

$\left [\begin{array}{} mL^2 && 0 \\ 0 && mL^2 \end{array} \right] \left[ \begin{array}{} \ddot{\theta}_1 \\ \ddot{\theta}_2 \end{array} \right] + \left[ \begin{array}{} mgL + ka^2 && - ka^2 \\ -ka^2 && mgL + ka^2 \end{array} \right] \left[ \begin{array}{} \theta_1 \\ \theta_2 \end{array} \right] = \left[ \begin{array}{} 0 \\ 0 \end{array} \right] $

$ \left[ \begin{array}{} mgL + ka^2 && - ka^2 \\ -ka^2 && mgL + ka^2 \end{array} \right] \rightarrow$ acoplamento estático

Considerando:

$\theta_1(t) = \Theta_1 cos (\omega t)$

$\theta_2(t) = \Theta_2 cos(\omega t) $

$-\omega^2 \left[ \begin{array}{} mL^2 && 0 \\ 0 && mL^2 \end{array} \right] \left[ \begin{array}{} \Theta_1 \\ \Theta_2 \end{array} \right] + \left[ \begin{array}{} mgL + ka^2 && -ka^2 \\ -ka^2 && mgL + ka^2 \end{array} \right] \left[ \begin{array}{} \Theta_1 \\ \Theta_2 \end{array} \right] = \left[ \begin{array}{} 0 \\ 0 \end{array} \right] $

Equação característica: 

$det \left[ \begin{array}{} mgL + ka^2 - \omega^2 m L^2 && - ka^2 \\ - k a^2 && mgL + ka^2 - \omega^2 m L^2 \end{array} \right] $

$ = \left( mgL + ka^2 - \omega^2mL^2 \right)^2 - \left(ka^2 \right)^2 = 0$

$mgL + ka^2 - \omega^2 m L^2 = \pm ka^2$

$\omega_1 = \sqrt{\dfrac{g}{L}} \qquad \omega_2 = \sqrt{\dfrac{g}{L} + 2 \dfrac{k}{m} \dfrac{a^2}{L^2}}$

Os modos de vibração nautrais são obtidos das equações:

$\left(-\omega_1^2 mL^2  + \left( mgL + ka^2 \right) \right) \Theta_1^{1} - ka^2 \Theta_2^{1} = 0$

$ -ka^2 \Theta^{2}_1 + \left( -\omega_2^2 mL^2  + mgL + ka^2 \right) \Theta^2_2  = 0$

Fazendo $\omega_1^2 = \dfrac{g}{L}$ e $\omega_2^2 = \dfrac{g}{L} + 2 \left( \dfrac{k}{m} \right) \left( \dfrac{a^2}{L^2} \right)$

$\dfrac{\Theta_2^{1}}{\Theta_1^{2}} = 1 \qquad \dfrac{\Theta_2^2}{\Theta_1^2} = -1 $

O movimento geral do sistema pode ser expresso como a superposição dos dois modos naturais multiplicados pelas coordenadas naturais associadas.

$\left[ \begin{array}{} \theta_1(t) \\ \theta_2() \end{array} \right] = C_1 \left[ \begin{array}{} \Theta_1^1 \\ \Theta_2^1 \end{array} \right] cos (\omega_1 t - \phi_1 ) + C_2 \left[ \begin{array}{} \Theta_1^2 \\ \Theta^2_2 \end{array} \right] cos ( \omega_2 t - \phi_2 ) $

Inserindo as razões de amplitude na equação anterior: 

$\theta_1(t) = C_1 cos (\omega_1 t - \phi_1) + C_2 cos ( \omega_2 t - \phi_2 ) $

$\theta_2(t) = C_1 cos (\omega_1 t - \phi_1) + C_2 cos (\omega_2 t - \phi_2 ) $



Considerando: $\theta_1(0) = \theta_0 \qquad \theta_2(0) = \dot{\theta}_1(0) = \dot{\theta}_2(0) = 0$

$\theta_1(t) = \dfrac{1}{2} \theta_0 cos \omega_1 t + \dfrac{1}{2} \theta_0 cos \omega_2 t = \theta_0 cos \left( \dfrac{\omega_2 - \omega_1}{2} t \right) cos \left( \dfrac{\omega_2 + \omega_1}{2} t \right)$

$\theta_2(t) = \dfrac{1}{2} \theta_0 cos (\omega_1 t) - \dfrac{1}{2} \theta_0 cos (\omega_2 t) = \theta_0 sen \left( \dfrac{\omega_2 - \omega_1}{2} t \right) sen \left( \dfrac{\omega_2 + \omega_1}{2} t \right)$

Escrevendo de modo mais compacto: 

$\theta_1(t) = \theta_0 cos \left( \dfrac{\omega_B t}{2} \right) cos ( \omega_{med} t )$

$\theta_2(t) = \theta_0 sen \left( \dfrac{\omega_B t}{2} \right) sen ( \omega_{med} t ) $

Onde:

$\dfrac{\omega_B}{2} = \dfrac{\omega_2 - \omega_1}{2} \approx \dfrac{1}{2} \dfrac{k}{m} \dfrac{a^2}{\sqrt{gL^3}}$

$\omega_B \rightarrow$ frequência de batimento

## Vibração Harmônica Forçada

```python
Image('figuras/fig5-9.png',width=600)
```

$k_1 + k_2 = k_{11} \qquad k_2 + k_3 = k_{22} \qquad -k_2 = k_{12} = k_{21}$

$\left[ \begin{array}{} m_1 && 0 \\ 0 && m_2 \end{array} \right] \left[ \begin{array} \ddot{x}_1 \\ \ddot{x}_2 \end{array} \right] + \left[ \begin{array}{} k_{11} && k_{12} \\ k_{12} && k_{22} \end{array} \right] \left[ \begin{array}{}  x_1 \\ x_2 \end{array} \right] = \left[ \begin{array}{} F_1 \\ 0 \end{array} \right] sen \omega t$ 

Supomos que a solução seja: 

$\left[ \begin{array}{} x_1(t) \\ x_2(t) \end{array} \right] = \left[ \begin{array}{} X_1 \\ X_2 \end{array} \right] sen \omega t$

Substituindo esta solução na primeira equação:

$\left[ \begin{array}{} \left( k_{11} - m_1 \omega^2 \right) && k_{12} \\ k_{12} && \left(k_{22} - m_2 \omega^2 \right) \end{array} \right] \left[ \begin{array}{} X_1 \\ X_2 \end{array} \right] = \left[ \begin{array}{} F_1 \\ 0 \end{array} \right] $

Ou:

$\left[Z(\omega)\right] \left[ \begin{array}{} X_1 \\ X_2 \end{array} \right] = \left[ \begin{array}{} F_1 \\ 0 \end{array} \right] $

$\left[ \begin{array}{} X_1 \\ X_2 \end{array} \right] = [Z(\omega)]^{-1} \left[ \begin{array}{} F_1 \\ 0 \end{array} \right] = \dfrac{adj[Z(\omega)}{det[Z(\omega)]}$



Observação: Matriz adjunta:

$A = \left[ \begin{array}{} a && b \\ c && d \end{array} \right] $

$adj(A) = \left[ \begin{array}{} d && -b \\ -c && a \end{array} \right] $

$det[Z(\omega)] = m_1 m_2 \left( \omega^2_1 - \omega^2 \right) \left( \omega_2^2 - \omega^2 \right)$

Onde $\omega_1$ e $\omega_2$ são as frequências dos modos normais

$\left[ \begin{array}{} X_1 \\ X_2 \end{array} \right] = \dfrac{1}{det[Z(\omega)]} \left[ \begin{array}{} \left(k_{22} - m_2 \omega^2 \right) && - k_{12} \\ -k_{12} && \left(k_{11} - m_{1} \omega^2 \right) \end{array} \right] \left[ \begin{array}{} F_1 \\ 0 \end{array} \right] $

$X_1 = \dfrac{\left( k_{22} - m_2 \omega^2 \right) F}{m_1 m_2 \left( \omega^2_1 - \omega^2 \right) \left( \omega_2^2 - \omega \right)}$

$X_2 = \dfrac{-k_{12}F}{m_1 m_2 \left( \omega_1^2 - \omega \right) \left( \omega_2^2 - \omega^2 \right)}$

## Absorvedor ed Vibração

Um sistema massa-mola $k_2$, $m_2$ afinado com a frequência da força excitadora tal que $omega^2 = \dfrac{k_2}{m_2}$, atuará como um absorvedor de vibração e reduzirá a zero o movimento da massa principal $m_1$. 

$\omega_{11}^2 = \dfrac{k_1}{m_1} \qquad \omega_{22}^2 = \dfrac{k_2}{m_2}$
